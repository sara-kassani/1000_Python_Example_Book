# Sigmoid

```python
def sigmoid_and_derivative(x):
    s = 1/(1+np.exp(-x))
    ds = s*(1-s)  
    return s, ds

x = np.arange(-6,6,0.01)
sigmoid_x, sigmoid_derivative_x = sigmoid_and_derivative(x)

fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x, sigmoid_x, color="b", linewidth=3, label="Sigmoid")
ax.plot(x, sigmoid_derivative_x, '--', color="r", linewidth=3, label="Derivative of Sigmoid")
ax.legend()
ax.grid(False)
fig.savefig('images/chapter_04_sigmoid_activation_function.png', dpi=300, bbox_inches='tight')
fig.savefig('images/chapter_04_sigmoid_activation_function.pdf', type='pdf', dpi=300, bbox_inches='tight')
```

# Hyperbolic Tangent

```python
def tanh_and_derivative(x):
    t = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt = 1-t**2
    return t, dt

x = np.arange(-4,4,0.01)
tanh_x, tanh_derivative_x = tanh_and_derivative(x)

fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x, tanh_x, color="b", linewidth=3, label="Hyperbolic Tangent (tanh)")
ax.plot(x, tanh_derivative_x, '--', color="r", linewidth=3, label="Derivative of tanh")
ax.legend()
ax.grid(False)
fig.savefig('images/chapter_04_tanh_activation_function.png', dpi=300, bbox_inches='tight')
fig.savefig('images/chapter_04_tanh_activation_function.pdf', type='pdf', dpi=300, bbox_inches='tight')
```

# ReLU

```python
def relu_and_derivative(x):
    r =  x * (x > 0)
    dr = 1. * (x > 0)
    return r, dr

x = np.arange(-2, 2, 0.01)
relu_x, relu_derivative_x = relu_and_derivative(x)

fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
# ax.spines['bottom'].set_position('bottom')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x, relu_x, color="b", linewidth=3, label="ReLU")
ax.plot(x, relu_derivative_x, '--', color="r", linewidth=3, label="Derivative of ReLU")
ax.legend()
ax.grid(False)
fig.savefig('images/chapter_04_relu_activation_function.png', dpi=300, bbox_inches='tight')
fig.savefig('images/chapter_04_relu_activation_function.pdf', type='pdf', dpi=300, bbox_inches='tight')
```
